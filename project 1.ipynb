```python
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_excel("/content/Dataset for Data Analytics.xlsx")

# Remove duplicates
df.drop_duplicates(inplace=True)

# Strip spaces from column names
df.columns = df.columns.str.strip()

# Clean data
for col in df.columns:
    if df[col].dtype in ['int64', 'float64']:
        df[col] = pd.to_numeric(df[col], errors='coerce')
        df[col].fillna(df[col].mean(), inplace=True)
    else:
        df[col] = df[col].astype(str).str.strip().str.lower()
        df[col].replace(['nan', 'none', ""], pd.NA, inplace=True)
        if not df[col].mode().empty:
            df[col].fillna(df[col].mode()[0], inplace=True)

# Outlier removal using IQR
Q1 = df.select_dtypes(include=['number']).quantile(0.25)
Q3 = df.select_dtypes(include=['number']).quantile(0.75)
IQR = Q3 - Q1
df = df[~((df.select_dtypes(include=['number']) < (Q1 - 1.5 * IQR)) |
          (df.select_dtypes(include=['number']) > (Q3 + 1.5 * IQR))).any(axis=1)]

# Final check
print("Remaining null values:\n", df.isnull().sum())

# Save cleaned file
df.to_excel('cleaned_data.xlsx', index=False)
print("Data cleaned and saved successfully!")
```
